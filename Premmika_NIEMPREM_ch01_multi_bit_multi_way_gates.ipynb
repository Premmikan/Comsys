## คัดลอกโค้ดของเกทจากแบบฝึกหัดที่แล้ว​

ให้คัดลอกนิยามอุปกรณ์สำหรับเกทในแบบฝึกหัดที่แล้ว (And, Or, Not, Xor, Mux และ DMux) มาวางไว้ในกล่องโค้ดด้านล่าง เพื่อให้สามารถนำเกทเหล่านี้มาใช้งานในแบบฝึกหัดนี้ได้

```
# Put your component codes from previous notebooks here
# DO NOT ERASE THIS CELL - to be graded

class Not(NotLayoutMixin, Component):
    IN = [w.In]
    OUT = [w.out]

    PARTS = [
        Nand(a = w.In, b = w.In , out = w.out)
    ]

# DO NOT ERASE THIS CELL - to be graded

class And(AndLayoutMixin, Component):
    IN = [w.a, w.b]
    OUT = [w.out]

    PARTS = [
        Nand(a = w.a,b=w.b, out = w.x),
        Not(In=w.x , out = w.out)
    ]

# DO NOT ERASE THIS CELL - to be graded

class Or(OrLayoutMixin, Component):
    IN = [w.a, w.b]
    OUT = [w.out]

    PARTS = [
        Not(In = w.a, out = w.x),
        Not(In = w.b , out = w.y),
        Nand(a = w.x, b = w.y , out = w.out)
    ]

# DO NOT ERASE THIS CELL - to be graded

class Xor(XorLayoutMixin, Component):
    IN = [w.a, w.b]
    OUT = [w.out]

    PARTS = [
        Not(In = w.b, out = w.x),
        And(a = w.x,b = w.a,out = w.y),
        Not(In = w.a , out = w.z),
        And(a = w.z , b = w.b , out = w.k),
        Or(a = w.y, b = w.k , out = w.out)
    ]

# DO NOT ERASE THIS CELL - to be graded

class Mux(Component):
    IN = [w.a, w.b, w.sel]
    OUT = [w.out]

    PARTS = [
        Not(In = w.sel , out = w.x),
        And(a = w.x , b = w.a , out = w.y),
        And(a = w.b , b = w.sel, out = w.z),
        Or(a = w.y, b = w.z, out = w.out)
    ]

# DO NOT ERASE THIS CELL - to be graded

class DMux(Component):
    IN = [w.In, w.sel]
    OUT = [w.a, w.b]

    PARTS = [
        Not(In = w.sel, out = w.x),
        And(a = w.In , b = w.x , out = w.a),
        And(a = w.In , b = w.sel , out = w.b)
    ]
```

## Multi-bit gates: `Not16`, `And16`, `Or16`, `Mux16`

ตัวอย่างการทำงานแบบโต้ตอบ
* [Not16 gate](https://ecourse.cpe.ku.ac.th/courses/comsys/demo/not16.html)
* [And16 gate](https://ecourse.cpe.ku.ac.th/courses/comsys/demo/and16.html)
* [Or16 gate](https://ecourse.cpe.ku.ac.th/courses/comsys/demo/or16.html)
* [Mux16 gate](https://ecourse.cpe.ku.ac.th/courses/comsys/demo/mux16.html)

ส่วนโค้ดสำหรับทดสอบจะอยู่ถัดจากการนิยามคลาสอุปกรณ์

```
# DO NOT ERASE THIS CELL - to be graded

class Not16(Component):
    IN = [w(16).In]
    OUT = [w(16).out]

    PARTS = [ # แนะนำให้สร้างเกท 16 ตัวด้วย for-loop

    ]
    for i in range(16):
      PARTS.append(Not(In = w.In[i], out = w.out[i]))
```

```
# DO NOT ERASE THIS CELL - to be graded

class And16(Component):
    IN = [w(16).a, w(16).b]
    OUT = [w(16).out]

    PARTS = [ # แนะนำให้สร้างเกท 16 ตัวด้วย for-loop

    ]
    for i in range(16):
      PARTS.append(And(a = w.a[i],b = w.b[i], out = w.out[i]))
```

```
# DO NOT ERASE THIS CELL - to be graded

class Or16(Component):
    IN = [w(16).a, w(16).b]
    OUT = [w(16).out]

    PARTS = [ # แนะนำให้สร้างเกท 16 ตัวด้วย for-loop
    ]
    for i in range(16):
      PARTS.append(Or(a = w.a[i],b = w.b[i] , out = w.out[i]))
```

```
# DO NOT ERASE THIS CELL - to be graded

class Mux16(Component):
    IN = [w(16).a, w(16).b, w.sel]
    OUT = [w(16).out]

    PARTS = [ # แนะนำให้สร้างเกท 16 ตัวด้วย for-loop
]
    for i in range(16):
      PARTS.append(Mux(a = w.a[i],b = w.b[i] , sel = w.sel, out = w.out[i]))
```

นิสิตสามารถทดลองและเทสเกทที่สร้างขึ้นโดยพิมพ์โค้ดสำหรับ trace, plot, interact ได้ที่ส่วนด้านล่างนี้

```
# พื้นที่ทดสอบ (ปรับเปลี่ยนตามต้องการ)

interact(Not16, depth=1)
interact(And16, depth=1)
interact(Or16, depth=1)
interact(Mux16, depth=1)
```

## Multi-way gates: `Or8Way`, `Mux4Way16`, `Mux8Way16`, `DMux4Way`, `DMux8Way`

ตัวอย่างการทำงานแบบโต้ตอบ
* [Or8Way](https://ecourse.cpe.ku.ac.th/courses/comsys/demo/or8way.html)
* [Mux4Way16](https://ecourse.cpe.ku.ac.th/courses/comsys/demo/mux4way16.html)
* [Mux8Way16](https://ecourse.cpe.ku.ac.th/courses/comsys/demo/mux8way16.html)
* [DMux4Way](https://ecourse.cpe.ku.ac.th/courses/comsys/demo/dmux4way.html)
* [DMux8Way](https://ecourse.cpe.ku.ac.th/courses/comsys/demo/dmux8way.html)

ส่วนโค้ดสำหรับทดสอบจะอยู่ถัดจากการนิยามคลาสอุปกรณ์

```
# DO NOT ERASE THIS CELL - to be graded

class Or8Way(Component):
    IN = [w(8).In]
    OUT = [w.out]

    PARTS = [Or(a = w.In[0], b = w.In[1], out = w.o1)]
    for i in range(1,6):
      PARTS.append(Or(a = w.In[i+1], b = getattr(w,f"o{i}"), out = getattr(w,f"o{i+1}")))
    PARTS.append(Or(a = w.In[7], b = w.o6, out = w.out))
```

```
# DO NOT ERASE THIS CELL - to be graded

class Mux4Way16(Component):
    IN = [w(16).a, w(16).b, w(16).c, w(16).d, w(2).sel]
    OUT = [w(16).out]

    PARTS = [
        Mux16(a = w.a, b = w.b, sel = w.sel[0], out = w(16).o1),
        Mux16(a = w.c , b = w.d, sel = w.sel[0], out = w(16).o2),
        Mux16(a = w.o1, b = w.o2, sel = w.sel[1], out = w.out)
    ]
```

```
# DO NOT ERASE THIS CELL - to be graded

class Mux8Way16(Component):
    IN = [
        w(16).a,
        w(16).b,
        w(16).c,
        w(16).d,
        w(16).e,
        w(16).f,
        w(16).g,
        w(16).h,
        w(3).sel,
    ]
    OUT = [w(16).out]

    PARTS = [
        Mux4Way16(a = w.a, b = w.b,c = w.c, d = w.d, sel = w.sel[0:2], out = w(16).o1),
        Mux4Way16(a = w.e, b = w.f, c = w.g, d = w.h, sel = w.sel[0:2], out = w(16).o2),
        Mux16(a = w.o1, b = w.o2, sel = w.sel[2], out = w.out)
    ]
```

```
# DO NOT ERASE THIS CELL - to be graded

class DMux4Way(Component):
    IN = [w.In, w(2).sel]
    OUT = [w.a, w.b, w.c, w.d]

    PARTS = [
        DMux(In = w.In, sel = w.sel[1], a = w.y, b = w.x),
        DMux(In = w.y, sel = w.sel[0], a = w.a, b = w.b),
        DMux(In = w.x , sel = w.sel[0], a = w.c, b = w.d)

    ]
```

```
# DO NOT ERASE THIS CELL - to be graded

class DMux8Way(Component):
    IN = [w.In, w(3).sel]
    OUT = [w.a, w.b, w.c, w.d, w.e, w.f, w.g, w.h]

    PARTS = [
        DMux(In = w.In, sel = w.sel[2], a = w.x, b = w.y),
        DMux4Way(In = w.x, sel = w.sel[0:2], a = w.a, b = w.b, c = w.c, d = w.d),
        DMux4Way(In = w.y, sel = w.sel[0:2], a = w.e, b = w.f, c = w.g, d = w.h)
    ]
```

ใช้พื้นที่ด้านล่างทดสอบความถูกต้องของการสร้างอุปกรณ์

```
# พื้นที่ทดสอบ (ปรับเปลี่ยนตามต้องการ)

interact(Or8Way, depth=1)
interact(Mux4Way16, depth=1)
interact(Mux8Way16, depth=1)
interact(DMux4Way, depth=1)
interact(DMux8Way, depth=1)
```
